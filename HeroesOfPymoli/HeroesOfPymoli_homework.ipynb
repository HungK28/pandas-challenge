```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data_df = pd.DataFrame(purchase_data)

total_player = purchase_data_df["SN"].count()
total_player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data_df["Item ID"].unique())
avg_purchase_price = round(purchase_data_df["Price"].mean(), 2)
total_purchases = purchase_data_df["Price"].count()
total_revenue = purchase_data_df["Price"].sum()

purchasing_analysis_df = pd.DataFrame([{"Number of Unique Items": unique_items, "Average Purchase Price": avg_purchase_price, 
                                        "Total Number of Purchases": total_purchases, "Total Revenue": total_revenue }])
purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = purchase_data_df["Gender"].value_counts()
gender_percentage = round(gender_count/ total_player * 100, 2)
gender_demo_df = pd.DataFrame({"Percentage of Players": gender_percentage, "Total Count": gender_count})

gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_count = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
gender_avg_purchase = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_total_purchase_value = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
gender_purchase_per_person = gender_total_purchase_value / gender_demo_df["Total Count"]

gender_analysis = pd.DataFrame({"Purchase Count": gender_purchase_count, "Average Purchase Price": round(gender_avg_purchase, 2), 
                                "Total Purchase Value": gender_total_purchase_value,  
                                "Avg Total Purchase per Person": round(gender_purchase_per_person, 2)})
gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 10, 14, 18, 23, 28, 33, 38, 99]
age_title =  ["<10", "10-14", "15-19", "20-24", "25,29", "30-34", "35-39", "40+"]
purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], age_bins, labels=age_title)
age_totals = purchase_data_df["Age Ranges"].value_counts() 
age_percents = age_totals/total_player*100

age_demo = pd.DataFrame({"Total Purchase Count": age_totals, "Percent of Players": round(age_percents, 2)})
age_demo.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], age_bins, labels=age_title)
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")
age_avg = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Count")
age_total_purchase = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
age_per_person = age_total_purchase/ age_demo["Total Purchase Count"]

age_analysis = pd.DataFrame({"Purchase Count": age_counts, "Average Purchase Price": age_avg, 
                             "Total Purchase Value": age_total_purchase, "Avg Total Purchase per Person": age_per_person})
age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
spender_avg = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Count")
spender_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

top_spenders_df = pd.DataFrame({"Purchase Count": spender_count, "Average Purchase Price": spender_avg,
                               "Total Purchase Value": spender_total})
spenders_sorted = top_spenders_df.sort_values("Total Purchase Value", ascending = False)
spenders_sorted.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_extracted = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]
items_count = purchase_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
items_avg = purchase_data.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Count")
items_total = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
item_data_pd = pd.DataFrame({"Purchase Count": items_count, "Item Price": round(items_avg, 2), "Total Purchase Value": items_total})

sorted_items = item_data_pd.sort_values("Purchase Count", ascending = False)
sorted_items.head(5)
```
